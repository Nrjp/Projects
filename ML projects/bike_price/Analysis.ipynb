# Exploratory Data Analysis of Bikes Prices Dataset


 - Maximum price,Mininum price Selling and Showroom price
 - Numbers of years in DataSet
 - Sales per years

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading Files
bike_df = pd.read_csv('BIKE DETAILS.csv')
```

```python
#Bike dataset information
print('Columns:',bike_df.shape[1])
print('Rows:',bike_df.shape[0])
print('\nFeatures:',bike_df.columns)
```

```python
#Printing Dataset
bike_df.head()
```

```python
# Null Values 
bike_df.isnull().sum()
```

## Selling price

```python
print("Statistics for Selling Price:\n")

print('Maximum price:      ₹',bike_df['selling_price'].max())
print('Mininum price:      ₹',bike_df['selling_price'].min())
print('Mean price:         ₹',bike_df['selling_price'].mean())
print('Median price:       ₹',bike_df['selling_price'].median())
print('Standard deviation:',bike_df['selling_price'].std())
```

```python
# Number of Unique Values in Name columns
bike_df['selling_price'].nunique()
```

```python
# plotting Selling price
sns.distplot(bike_df['selling_price'])
```

### Showroom Price 


```python
print("Statistics for Showroom Prices:\n")

print('Maximum price:      ₹',bike_df['ex_showroom_price'].max())
print('Mininum price:      ₹',bike_df['ex_showroom_price'].min())
print('Mean price:         ₹',bike_df['ex_showroom_price'].mean())
print('Median price:       ₹',bike_df['ex_showroom_price'].median())
print('Standard deviation:',bike_df['ex_showroom_price'].std())
```

```python
# Number of Unique Values in Name columns
bike_df['ex_showroom_price'].nunique()
```

```python
#Plotting Showroom Price
sns.distplot(bike_df['ex_showroom_price'])
```

### Plotting Values

```python
# PLotting Scatter plot of Selling price and showroom price
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.regplot(bike_df['selling_price'],bike_df['ex_showroom_price'],color='green')
plt.xlabel('Selling Prices')
plt.ylabel('Showroom Prices')
plt.show()
```

```python
# plotting Showroom Price,Selling Price and difference  
plt.rcParams['figure.figsize'] = 15,8

df1 = bike_df.copy()
# set width of bar
barWidth = 0.2
# Difference in price
diff_prices = df1['ex_showroom_price']- df1['selling_price']

# set height of bar
bars1 = df1['ex_showroom_price'].head(20)
bars2 = df1['selling_price'].head(20)
bars3 = diff_prices.head(20)
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='green', width=barWidth, edgecolor='white', label='Showroom Price')
plt.bar(r2, bars2, color='blue', width=barWidth, edgecolor='white', label='Selling Price')
plt.bar(r3, bars3, color='red', width=barWidth, edgecolor='white', label='diff')
 
# Add xticks on the middle of the group bars
plt.xticks([r + barWidth for r in range(len(bars1))],[n for n in range(0, 20)])
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
