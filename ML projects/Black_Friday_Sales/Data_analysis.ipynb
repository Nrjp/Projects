### User ID

```python
# Numbers of Users
print('Total number of Users:',df['User_ID'].nunique())
```

```python
# users
users = df['User_ID'].value_counts()
```

```python
# Top 10 Users
print('Top 10 user_id:\n',users.head(10))
```

```python
# Numbers of users above 500 Purchase
print('User ID above 500 Purchase:',users[users>500].count())
# Numbers of users above 500 Purchase
print('User ID below 500 and above 100 Purchase:',users[(users<500)&(users>100)].count())
# Numbers of users below 100 Purchase
print('User ID Below 100 Purchase:',users[users<=100].count())
# Numbers of users having 10 Purchase
print('User ID having 10 Purchase:',users[users==10].count())
```

###  Products

```python
# Number of Products
print('Numbers of Products:',df['Product_ID'].nunique())
```

```python
# Products
products = df['Product_ID'].value_counts()
```

```python
# Top 10 sale products
print('Top 10 Product_ID:\n',products.head(10))
```

```python
# Low sale products
print('Top 10 Product_ID:\n',products.tail())
```

```python
# Numbers of products above 500 Purchase
print('Product_ID above 500 Purchase:',products[products>500].count())
# Numbers of products above 500 Purchase
print('Product_ID below 500 and above 100 Purchase:',products[(products<500)&(products>100)].count())
# Numbers of products below 100 Purchase
print('Product_ID Below 100 Purchase:',products[products<=100].count())
```
