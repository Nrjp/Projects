# Analysis

```python
#Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading Datafile
df  = pd.read_csv("heart.csv")

print('Rows:',df.shape[0])
print('Columns:',df.shape[1])
print('\nFeatures:',df.columns)
```

### Target Values

```python
# cheacking Target Values
df['target'].value_counts()
```

```python
#Anlysis of Target Variable
sns.countplot(df['target'])
```

```python
# Getting Percentage of Target
healthy = ((len(df[df.target == 0]) / (len(df.target))*100))
conditions = ((len(df[df.target == 1]) / (len(df.target))*100))
print('Patiants Not having Heart Conditions: {:.2f}%'.format(healthy))
print('Patiants having Heart Conditions: {:.2f}%'.format(conditions))
```

### Age Values

```python
#Anlysis of Age
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.countplot(df['age'])
```

### Sex

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.countplot(df['sex'])
plt.xlabel("Sex (0 = female, 1= male)")
plt.show()
```

```python
df['sex'].value_counts()
```

```python
# Percentage of Male and Feamale
healthy = ((len(df[df.target == 0]) / (len(df.target))*100))

female = ((len(df[df.sex==0])/ (len(df.sex)) *100))
male   = ((len(df[df.sex==1])/ (len(df.sex)) *100))

print('Percentage of Female Patients:{:.2f}%'.format(female))
print('Percentage of male Patients:{:.2f}%'.format(male))
```

```python
pd.crosstab(df.age,df.target).plot(kind="bar",figsize=(20,6))
plt.title('Heart Disease by Ages')
plt.xlabel('Age')
plt.ylabel('Targets')
```

```python
pd.crosstab(df.sex,df.target).plot(kind="bar",figsize=(20,6))
plt.title('Heart Disease by Sex')
plt.xlabel("Sex (0 = female, 1= male)")
plt.ylabel('Targets')
```
